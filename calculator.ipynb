```python
from tkinter import *
from tkinter.font import Font
root=Tk()
root.configure(bg="black")
root.title("Calculator")
root.resizable(width = False, height=False)
w=411
h=405
sw=root.winfo_screenwidth()
sh=root.winfo_screenheight()
x=(sw/2)-(w/2)
y=(sh/2)-(h/2)
root.geometry("%dx%d+%d+%d"% (w,h,x,y))
root.iconbitmap('favicon.ico')
e=Entry(root,width=63,borderwidth=16,bg="black",fg="white")
e.grid(row=0,column=0,columnspan=4)

def btnClick(number):
    current=e.get()
    e.delete(0,END)
    e.insert(0,str(current)+str(number))

def clear():
    e.get()
    e.delete(0,END)

def btndiv():
    global math,Fnum
    math=1
    firstnum=e.get()
    Fnum=int(firstnum)
    e.delete(0,END)
def btnMul():
    global math,Fnum
    math=2
    firstnum=e.get()
    Fnum=int(firstnum)
    e.delete(0,END)
def btnAdd():
    global math,Fnum
    math=3
    firstnum=e.get()
    Fnum=int(firstnum)
    e.delete(0,END)  
def btnSub():
    global math,Fnum
    math=4
    firstnum=e.get()
    Fnum=int(firstnum)
    e.delete(0,END) 

def equal():
    secondnum=e.get()
    e.delete(0,END)
    if math==1:
        e.insert(0,Fnum/int(secondnum))
    elif math==2:
        e.insert(0,Fnum*int(secondnum))
    elif math==3:
        e.insert(0,Fnum+int(secondnum))
    else:
        e.insert(0,Fnum-int(secondnum))

buttonZero=Button(text="0",fg="white",bg="black",padx=40,pady=20,font=("times",13,"bold"),command=lambda:btnClick(0)).grid(row=1,column=0)
buttonDiv=Button(text="/",fg="white",bg="black",padx=40,pady=20,font=("times",13,"bold"),command=btndiv).grid(row=1,column=1)
buttonMul=Button(text="*",fg="white",bg="black",padx=40,pady=20,command=btnMul,font=("times",13,"bold")).grid(row=1,column=2)
buttonAdd=Button(text="+",fg="white",bg="black",padx=40,pady=20,command=btnAdd,font=("times",13,"bold")).grid(row=1,column=3)
button1=Button(text="1",fg="black",bg="orange",padx=40,pady=20,font=("times",13,"bold"),command=lambda:btnClick(1)).grid(row=3,column=0)
button2=Button(text="2",fg="black",bg="orange",padx=40,pady=20,font=("times",13,"bold"),command=lambda:btnClick(2)).grid(row=3,column=1)
button3=Button(text="3",fg="black",bg="orange",padx=40,pady=20,font=("times",13,"bold"),command=lambda:btnClick(3)).grid(row=3,column=2)
buttonSub=Button(text="-",fg="white",bg="black",padx=42,pady=20,font=("times",13,"bold"),command=btnSub).grid(row=3,column=3)
button4=Button(text="4",fg="black",bg="orange",padx=40,pady=20,font=("times",13,"bold"),command=lambda:btnClick(4)).grid(row=4,column=0)
button5=Button(text="5",fg="black",bg="orange",padx=40,pady=20,font=("times",13,"bold"),command=lambda:btnClick(5)).grid(row=4,column=1)
button6=Button(text="6",fg="black",bg="orange",padx=40,pady=20,font=("times",13,"bold"),command=lambda:btnClick(6)).grid(row=4,column=2)
buttonEqual=Button(text="=",padx=40,pady=90,fg="white",bg="black",font=("times",13,"bold"),command=equal).grid(row=4,column=3,rowspan=3)
button7=Button(text="7",fg="black",bg="orange",padx=40,pady=20,font=("times",13,"bold"),command=lambda:btnClick(7)).grid(row=5,column=0)
button8=Button(text="8",fg="black",bg="orange",padx=40,pady=20,font=("times",13,"bold"),command=lambda:btnClick(8)).grid(row=5,column=1)
button9=Button(text="9",fg="black",bg="orange",padx=40,pady=20,font=("times",13,"bold"),command=lambda:btnClick(9)).grid(row=5,column=2)
buttonClear=Button(text="Clear",fg="white",bg="black",padx=126,pady=20,font=("times",13,"bold"),command=clear).grid(row=6,column=0,columnspan=3)

root.mainloop()
```
